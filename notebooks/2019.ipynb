# 2019 - Singapore Weather Analysis

```python
# Import Pandas
import pandas as pd

# Add csv file
df = pd.read_csv("../dataset/weather_clean.csv")
```

We only care about the data from the year 2019, so i'll filter out to only get that

```python
# Filtering out data to only get 2019
df = df[df["Year"] == 2019]
```

## Max temperature & Max wind speed

For a good looking metric card, recording the max temperature & wind speed and which date it's on

```python
# Get the row with the hottest day
hottest_day_row = df.loc[df["Maximum Temperature (°C)"].idxmax()]

# Hottest date & Temperature
hottest_date = hottest_day_row["Date"]
hottest_temp = hottest_day_row["Maximum Temperature (°C)"]

# Get the row with the windiest day
windiest_day_row = df.loc[df["Max Wind Speed (km/h)"].idxmax()]

# Windiest date & speed
windiest_date = windiest_day_row["Date"]
windiest_temp = windiest_day_row["Max Wind Speed (km/h)"]
```
