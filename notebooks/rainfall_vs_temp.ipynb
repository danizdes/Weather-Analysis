# Total Rainfall vs Temperature

This part aims to test the hypothesis: Does more total rainfall lead to lower maximum temperatures?

```python
# Import pandas
import pandas as pd

# Import Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Add CSV file
df = pd.read_csv("../dataset/weather_clean.csv")
```

We only need the rainfall and the maximum temperatures soo:

```python
# Filtering out data to only get rainfall and maximum temperature
df = df[["Daily Rainfall Total (mm)", "Maximum Temperature (°C)"]]
```

Now, we need to assign what qualifies as **"rainy"** and **"dry"**. For this experiment, anything greater than 5mm would be considered rainy. 

```python
# Categorizing Days
df["Rainy Day"] = df["Daily Rainfall Total (mm)"].apply(lambda x: "Rainy" if x > 5 else "Dry")
```

Create the graph next

```python
# Create the figure
fig, ax = plt.subplots(figsize=(15,9))
```

```python
# Plot the graph
sns.boxplot(
    data=df,
    x="Rainy Day",
    y="Maximum Temperature (°C)",
    palette={"Rainy": "#3498db", "Dry": "#e67e22"},
    ax=ax
)
```

```python
# Save the plot
fig.savefig(fname="../media/rainfall_vs_temp.png", format="png", dpi=300, bbox_inches="tight")
```
