# Rainfall vs Wind Speed

Here I want to test a different hypothesis: Major rain events are more likely to be followed by wind gusts.

```python
#import pandas
import pandas as pd

# Import Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

#adding CSV file
df = pd.read_csv("../dataset/weather_clean.csv")
```

Now we only need rainfall and max wind speed, so let's filter that out

```python
df = df[["Daily Rainfall Total (mm)", "Max Wind Speed (km/h)"]]
```

I've done this like a bajillionn times, now let's make the plot

```python
# Create graph
fig, ax = plt.subplots(figsize=(10, 6))
```

```python
# Plot graph
sns.regplot(
    data=df,
    x="Daily Rainfall Total (mm)",
    y="Max Wind Speed (km/h)",
    scatter_kws={"alpha": 0.6, "color": "#226d9f", "edgecolor": "w", "s": 40},
    line_kws={"color": "#d63d2d", "linewidth": 2.3},
    ax=ax
) 
```

```python
# Set plot details
ax.set_title("Daily Rainfall vs Maximum Wind Temperature", fontsize=14)
ax.grid(True, linestyle="--", alpha=0.3)
```

```python
# Save the plot
fig.savefig(fname="../media/rainfall_vs_wind_speed.png", format="png", dpi=300, bbox_inches="tight")
```
